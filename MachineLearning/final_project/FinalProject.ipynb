- Operação

 - eu coloco na **setagem da máquina** quantas feições eu achar que a máquina deve trabalhar (ex: **POI** e **Salary**)

   - daí eu alimento a primeira ficha na máquina. Por exemplo, estas fichas contém a foto de uma pessoa e um rótulo True (é um homem). OK, aprendeu...

## Seleção de feições

**Desempenho** de uma máquina de **Machine Learning** tem a ver com algumas características e estratégias, como:

- se eu soube eliminar/tratar adequadamente **Outliers** antes de enfiar na minha máquina os dados

- tipo da máquina escolhida (se é uma **Naïve Bayes**, uma **Regressão**, uma **KNN** (K-Nearest Neighborhood), uma **SVM** (support Vector Machine), uma **Decision Tree**, uma **Random Forest**, uma **Adaboost**) 

- quantas **feições** eu escolhi (poucas feições, a máquina trabalha precariamente, feições demais e a máquina começa a ser contaminada por **ruído**... além disso, eu posso **combinar** duas feições em uma nova, híbrida...

- se eu fiz algum tipo de **preparação** dos dados:

 - eliminar ou tratar os **outliers** de alguma maneira

- se eu **tratei** adequadamente minhas feições, fiz:

 - o **escalonamento** - aquele negócio de criar as ascalas $0 \rightarrow 1$
 
  - o **PCM** - aquele negócio do girar e centralizar

- se eu não **apertei** ou **afrouxei** demais o processo de aprendizado da máquina (justa demais, a classificação fica **tosca** e frouxa demais, fica complexa demais e totalmente inadequada para interpretar **novos dados**)

- detalhes de setagem e submétodos próprios de cada máquina, Tipo: 

 - Naïve Bayes $\rightarrow$ **Gaussiana**
 
 - Regressão $\rightarrow$ **Linear**, **Logística**, **Logarítimica**
 
 - SVM $\rightarrow$ **Linear**, **Não-linear**
 
 - KNN $\rightarrow$ quantos **elementos** a considerar para a vizinhança, separação **Linear** ou **Não-linear**
 
 - mecanismo de busca do ponto ótimo por **Gradient Descend**

- a **lentidão** do meu processo geral (uma ótima máquina poderia levar virtualmente **milhares de anos** para ser treinada - ou se nos fiarmos ao **Guia do Mochileiro das Galáxias**, em **milhões de anos**

#### Primeira - redução de feições

O **PCA** é usado para transformar pares de **feições** em uma **componente principal** (PC). E esta sera usada como uma nova feição:

- A segunda dimensão (menor) será considerada o **ruído**

- Na **redução de dimensionalidade** (encontra **feições latentes**)

- Os algoritmos normalmente trabalham melhor com menos feições

 - whiten - se ligado, os vetores são multiplicados pela raiz quadra de n-amostras e divididos pelos valores singulares. Ele acaba removendo alguma informação, como as escalas relativas dos componentes, mas às vezes aumenta a acuidade preditiva de estimadores abaixo deles
 
 -svd_solver - é o tipo de solver utilizado (aleatório, arpak, automático, total)

Whitening will remove some information from the transformed signal (the relative variance scales of the components) but can sometime improve the predictive accuracy of the downstream estimators by making their data respect some hard-wired assumptions.

### O escolhido foi: Adaboost

#### Como funciona

O **Adaboost** trabalha normalmente tendo como base de montagem **Decision Tree**. Será formada uma floresta (um pouco como **Random Forest**), mas dessa vez baseada em **tocos**. Um toco em essência é um ramo com duas folhas (max_depth=1).

- Normalmente árvores classificadoras baseados em uma única divisão oferecem resultados muito **ruins** (dor no peito $\rightarrow$ infarto). **Tocos** são considerados **weak learners**. Mas a concepção do **Adaboost** exige isso (afinal ele trabalha com uma **floresta** desses **tocos** para fazer suas classificações!)

- Em **Decision Tree**, cada árvore possui o mesmo peso de voto (é uma **democracia plena**), mas agora cada um desses **tocos** possui um poder de voto **mutável** (e a cada iteração ele pode ser ajustado)

- Em **Decision Tree**, não importa qual a **ordem de decisão**. Os erros que o **primeiro toco** cometeu influenciam qual o **poder de voto** que terá o **segundo toco** e assim em diante

A floresta de tocos é então criada baseada na selação **simples** das minhas feições e cada toco ganha um peso amostral inicial. Este é dado simplesmente se dividindo 1 pelo número de tocos da minha floresta. Como **importa** a ordem de decisão, então o primeiro toco a ser mobilizado será aquele com o **menor índice Gini** (ou o critério colocado no seu lugar). Esse critério indica que aquela feição foi a que até agora melhor explicou individualmente (o que é uma decisão **fraca**!) o rótulo. No entanto, de início, os **poderes de voto** de cada toco são iguais. 

Em seguida, a partir do primeiro toco escolhido, o **poder de voto** do toco é ajustado a partir de uma adaptação **logística** do total de erros cometidos:

- um toco que não é melhor do que um lançamento de moeda terá o seu **poder de voto** igual a **Zero**

- um toco que prediz relativamente bem, acaba ganhando um **poder de voto** com um número **positivo** bastante elevado

- um toco que prediz o contrário do que deveria, acaba ganhando um **poder de voto** com um número **negativo** bastante elevado

*Como o **Erro Total** igual a **Zero** ou **Um** arruína a equação, é introduzido um **pequeno ruído** para que isso não aconteça*

Ao ser ajustado, um toco muda a **ponderação de voto** de todos os que vierem a seguir dele (pois esta sempre somará **Um**): 

- se um toco cometeu muitos **erros**, nós enfatizamos que os outros devem compensar o erro cometido por aquele, aumentando o peso de voto dos demais e **diminuindo** o seu poder de voto!

- se um toco se sair bastante **bem** nos resultados, nós aumentamos o seu **poder de voto** a partir de uma fórmula específica exponencial e **diminuímos** o poder de voto dos demais!

*As fórmulas de ajuste incluem passos **pequenos**, assim o algoritmo caminha devagar para evitar dar voz apenas a um **pequeno toco ditador**!*

E então, nós fazemos uma nova **iteração**, com nova classificação de **ordem dos tocos** (segundo o menor valor **Gini**) e continuamos o processo até que na nova iteração os ajustes passem a ser insignificantes.

*Observe que o índice **Gini** coloca muita força sobre a classificação **correta**! Isso é bom por exemplo em testes de medicamentos? Sem dúvida! E para o nosso caso? Epa, parece não ser nada bom! Vamos ver...*

(isso **não** turbina nossa máquina!)

- **random_state** (default=None) - a semente aleatória. normalmente usamos (random_state=42) para que o avaliador possa ver o que fizemos da mesma maneira que nós. **Manteremos nosso padrão** (isso **não** turbina nossa máquina!)

## Validação e Avaliação


### Métricas de avaliação

Quando estamos fazendo uma validação **Estatística** de um modelo, estamos preocupados com a quantidade de erros do **Tipo I** ("falso positivo") e do **Tipo II** ("falso negativo"). Neste tipo de análise, os mais perigosos normalmente são os erros do **Tipo I** e a análise é fortemente centrada nisso.

Estatísitca é uma ciência bastante **recente** e para isso, foram elaboradas diversas métricas de avaliação dos resultados. Não iremos entrar em detalhes aqui. Mas os dois grandes números para a **Validação de Modelos** são **Especificidade** e **Variabilidade**.

Em **Machine Learning**, embora utilizemos procedimentos muito análogos, a ciência ao evoluir, foi se destacando como uma espécie de **forking** da **Validação de Modelos**. E os novos cientistas perceberam que as ferramentas tradicionais não se adequavam muito bem ao trabalho deles. Então foram criados **novas métricas**, mais adequadas ao trabalho de:

- **Primeiro**, ensinar (treinar) uma máquina algo, através de rótulos (**Supervised Learning**) ou sem rótulos (**Unsupervised Learning**)

- **Segundo**, dar novos dados a esta máquina, diferentes daqueles nas quais ela foi treinada e assegurar que ela está fazendo um bom trabalho

Assim, a preocupação maior não é afastar o fantasma do **falso positivo** e sim verificar se a minha máquina está fazendo um trabalho razoável... As métricas mais utilizadas para avaliar esta performance geral (estamos trabalhando com algo **complexo**) utilizadas atualmente são:

- **Accuracy** (acuidade) - é um termo complexo e que envolve basicamente o conceito de **perspicácia**. Normalmente se compara a acuidade à ponta de um lápis: um lápis mal apontado tratrá um traço grosso com grande margem de dubiedade (a namorada flagrada no escuro, com uma silhueta de um homem - estaria sendo roubada, seria uma traição?)

- **Recall** (revocação) - o termo em português não ajuda muito. Mas é basicamente a máquina conseguir identificar a figura que lhe é mostrada (ou o **padrão** que lhe é apresentado). Então este padrão aqui se parece muito bem com um POI, ou aquele rosto com o de George W. Bush... uma máquina incapaz de **revocar** é como um cego, que posta uma figura à sua fente, nada diz

- **Precision** (precisão) - Esta é a capacidade da máquina, nas figuras que ela identificou, identificar **corretamente**. Então identificar George W. Bush, quando de fato foi lhe mostrada uma figura de George W. Bush, etc..

*Estas coisas são análogos aos testes aplicados em seres humanos.* 

- **F1** - uma combinação harmônica das duas anteriores. Quando queremos apenas um número para falar da capacidade da nossa máquina, este serve.

```python
import pickle
import sys
import numpy as np
from sklearn.model_selection import StratifiedShuffleSplit
sys.path.append("c:/udacity/udamini/tools/")
from feature_format import featureFormat, targetFeatureSplit

PERF_FORMAT_STRING = "\
\tAccuracy: {:>0.{display_precision}f}\tPrecision: {:>0.{display_precision}f}\t\
Recall: {:>0.{display_precision}f}\tF1: {:>0.{display_precision}f}\tF2: {:>0.{display_precision}f}"
RESULTS_FORMAT_STRING = "\tTotal predictions: {:4d}\tTrue positives: {:4d}\tFalse positives: {:4d}\
\tFalse negatives: {:4d}\tTrue negatives: {:4d}"

def test_classifier(clf, dataset, feature_list, folds=1000):
    data = featureFormat(dataset, feature_list, sort_keys=True)
    labels, features = targetFeatureSplit(data)
    labels = np.array(labels)
    features = np.array(features)
    sss = StratifiedShuffleSplit(n_splits=folds, random_state=42)
    print sss
    sss.get_n_splits(features, labels)
    print sss
    true_negatives = 0
    false_negatives = 0
    true_positives = 0
    false_positives = 0
    for train_index, test_index in sss.split(features, labels):
        features_train, features_test = features[train_index], features[test_index]
        labels_train, labels_test = labels[train_index], labels[test_index]        
        ### fit the classifier using training set, and test on test set
        clf.fit(features_train, labels_train)
        predictions = clf.predict(features_test)
        for prediction, truth in zip(predictions, labels_test):
            if prediction == 0 and truth == 0:
                true_negatives += 1
            elif prediction == 0 and truth == 1:
                false_negatives += 1
            elif prediction == 1 and truth == 0:
                false_positives += 1
            elif prediction == 1 and truth == 1:
                true_positives += 1
            else:
                print "Warning: Found a predicted label not == 0 or 1."
                print "All predictions should take value 0 or 1."
                print "Evaluating performance for processed predictions:"
                break
    try:
        total_predictions = true_negatives + false_negatives + false_positives + true_positives
        accuracy = 1.0*(true_positives + true_negatives)/total_predictions
        precision = 1.0*true_positives/(true_positives+false_positives)
        recall = 1.0*true_positives/(true_positives+false_negatives)
        f1 = 2.0 * true_positives/(2*true_positives + false_positives+false_negatives)
        f2 = (1+2.0*2.0) * precision*recall/(4*precision + recall)
        print clf
        print PERF_FORMAT_STRING.format(accuracy, precision, recall, f1, f2, display_precision = 5)
        print RESULTS_FORMAT_STRING.format(total_predictions, true_positives, false_positives, false_negatives, true_negatives)
        print ""
    except:
        print "Got a divide by zero when trying out:", clf
        print "Precision or recall may be undefined due to a lack of true positive predicitons."

CLF_PICKLE_FILENAME = "my_classifier.pkl"
DATASET_PICKLE_FILENAME = "my_dataset.pkl"
FEATURE_LIST_FILENAME = "my_feature_list.pkl"

def dump_classifier_and_data(clf, dataset, feature_list):
    with open(CLF_PICKLE_FILENAME, "w") as clf_outfile:
        pickle.dump(clf, clf_outfile)
    with open(DATASET_PICKLE_FILENAME, "w") as dataset_outfile:
        pickle.dump(dataset, dataset_outfile)
    with open(FEATURE_LIST_FILENAME, "w") as featurelist_outfile:
        pickle.dump(feature_list, featurelist_outfile)

def load_classifier_and_data():
    with open(CLF_PICKLE_FILENAME, "r") as clf_infile:
        clf = pickle.load(clf_infile)
    with open(DATASET_PICKLE_FILENAME, "r") as dataset_infile:
        dataset = pickle.load(dataset_infile)
    with open(FEATURE_LIST_FILENAME, "r") as featurelist_infile:
        feature_list = pickle.load(featurelist_infile)
    return clf, dataset, feature_list

def main():
    clf, dataset, feature_list = load_classifier_and_data()
    test_classifier(clf, dataset, feature_list)
    
main()
```
