### Consistência e preparação dos dados

#### Foi recebido:


- Um arquivo em formato **.csv** contendo dados de Twitagem do grupo, tendo já sido feito trabalho de elaboração sobre ele

 - nomes de cães foram capturados
 
 - campos de uma maneira geral já apresentavam alguma formatação demonstrando os objetivos a serem cumpridos
 
 
- Um caminho a um conjunto de fotos a serem baixadas

 - neste foi feito um trabalho de **Machine Learning** a fim de interpretar, em cada foto, qual o cão que a aprecia
 
 - acompanhou um dataset onde é possível extrair o caminho para cada foto
 
 
- A indicação do endereço do grupo no **Twitter**


#### Problemas identificados:

O arquivo principal, o **.csv** aparentava melhor **qualidade** do que o que posteriormente se percebeu:


- Problemas de qualidade identificados

 - o campo **tweet_id** encontrava-se formatado como Inteiro Longo e não como String de Inteiros

 - o campo **timestamp** não seguia o formato Date
 
 - o campo **name** continha muitos registros de nomes inválidos
 
 - os campos **puppo**, **pupper**, **doggo** e **floofer** encontravam-se incompletos
 
 - os campos **rating_numerator** e **rating_denominator** continham entradas inválidas
 
 - linhas de registro de **retwitagens**, inúteis para o projeto
 
 - linhas de registo de **retwiter**, inúteis ao projeto
 
 - linhas de registro sem fotografias, inúteis ao projeto
 

O arquivo baixado do Twitter, também apresentava alguns problemas de **qualidade**, a saber:


- Problemas de qualidade identificados  
 
 - o campo **id** possuía nome diferente de **tweet_id**
 
 - o campo **id** apresentava-se formatado como Inteiro Longo, como o anterior
 
 - o campo **retweet** estava formatado como Float, sendo Inteiro
 
 - o campo **favorites** estava formatado como Float, sendo Inteiro
 
 
Tendo como referência o arquivo **.csv**, sendo ele o objeto deste trabalho, alguns problemas de **tidyness** foram identificados, a saber:


 - o campo **retweet** se encontrava no arquivo vindo do servidor Twitter
 
 - o campo **favorites** se encontrava no mesmo arquivo descrito acima
 
 - o campo **link** com a foto dos cães, se encontrava no arquivo vindo do trabalho de **Machine Learning**

### Metodologia Empregada


Foi usada a técnica baseada em **Iterações** de ciclos de aquisição, verificação e correção, até em tese se obter dados perfeitos, ou aperfeiçoados, como no caso. Ela é uma metologia baseada em três etapas:

- **Gather**, na qual se **enlaça** ou se captura, a maior quantidade de dados novos potencialmente úteis ao projeto

- **Assess**, na qual se **esmiúça** os dados capturados, na tentativa de encontrar possíveis erros ou inconsistências, propondo soluções

- **Clean**, na qual se **elimina** ou **repara** os problemas encontrados


Para o projeto em questão, foram feitas **duas Iterações** de **GAC**, tendo como resultado, um dataset **melhorado**

### Propostas de Solução

Para o arquivo principal **.csv**, com relação à **qualidade**:


- Proposta de solução

 - o campo **tweet_id** ser convertido para String de Inteiros

 - o campo **timestamp** ser convertido para Date
 
 - o campo **name** ter nomes inválidos removidos e ser realizada nova busca por nomes no texto
 
 - os campos **puppo**, **pupper**, **doggo** e **floofer** completar estes campos, trazendo informações do texto
 
 - os campos **rating_numerator** e **rating_denominator** serem reparados quando possível, buscando dados no texto, ou na lógica dos campos ou quando não for possível, deixados em branco
 
 - linhas de registro de **retwitagens**, eliminar estas linhas
 
 - linhas de registo de **retwiter**, eliminar estas linhas
 
 - linhas de registro sem fotografias, eliminar estas linhas
 

O arquivo baixado do Twitter, com relação à **qualidade**:


- Proposta de solução  
 
 - o campo **id** ser corrigido para **tweet_id**
 
 - o campo **id** ser convertido para String de Inteiros
 
 - o campo **retweet** ser convertido para Inteiro
 
 - o campo **favorites** ser convertido para Inteiro
 
 
Tendo como referência o arquivo **.csv**, sendo ele o objeto deste trabalho, alguns problemas de **tidyness** foram identificados, a saber:

- Proposta de solução

 - o campo **retweet** ser trazido para o **.csv**
 
 - o campo **favorites** ser trazido para o **.csv**
 
 - o campo **link** ser trazido para o **.csv**
